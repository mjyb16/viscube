# Beginner's Guide: Gridding a Spectral Cube with VisCube 

In this tutorial, we will be using VisCube to grid a spectral cube. For simplicity, this tutorial skips the somewhat arduous process of extracting data from Measurement Sets into memory/numpy (for how to do that using CASA or XRADIO, see [here](https://viscube.readthedocs.io/en/latest/extracting_ms_to_py.html)). This tutorial focuses on JUST VisCube's gridding abilities themselves. For simplicity, we will be doing UV gridding (assuming the array is planar, so no W coordinates, a reasonable assumption for e.g. ALMA). 

**To demonstrate VisCube's power, we will be calculating the gridded visibilities AND the associated standard deviations for a spectral cube of galaxy NGC4697!**

```python
path = "/home/darthbarth/Big_red/NGC4697/cube_extracted_antennae.npz"
data = np.load(path)
freq = data["chan_freq_hz"]
u_raw = data["u"]
v_raw = data["v"]
vis_raw = data["vis"]
weight_raw = data["weight"]
sigma_re_cube = data["sigma_re_cube"]
sigma_im_cube = data["sigma_im_cube"]
mask = data["mask"]
```

```python
# Mask + compact
frequencies, u0, v0, vis0, weight0, sigma_re0, sigma_im0 = load_and_mask_with_sigma(
    freq, u_raw, v_raw, vis_raw, weight_raw, sigma_re_cube, sigma_im_cube, mask
)
```

```python
# Hermitian augment (including sigma)
uu, vv, vis_re, vis_imag, wt, sigma_re_aug, sigma_im_aug = hermitian_augment_with_sigma(
    u0, v0, vis0, weight0, sigma_re0, sigma_im0
)
```

```python
# Convert sigma -> inverse variance for fallback branch
eps_sigma = 1e-12
invvar_group_re = 1.0 / np.maximum(sigma_re_aug, eps_sigma)**2
invvar_group_im = 1.0 / np.maximum(sigma_im_aug, eps_sigma)**2
```

```python
# Grid
std_min_effective = 10
mean_re, mean_im, std_re, std_im, counts, u_edges, v_edges = grid_cube_all_stats_antenna_noise(
    frequencies=freq,
    uu=uu,
    vv=vv,
    vis_re=vis_re,
    vis_imag=vis_imag,
    weight=wt,  # your usual weights for mean + empirical std branch
    invvar_group_re=invvar_group_re,
    invvar_group_im=invvar_group_im,
    npix=501,
    window_name="kaiser_bessel",
    window_kwargs={"m": 6},
    std_min_effective=std_min_effective,
)
```

```python
mean_re_0, mean_im_0, std_re_0, std_im_0, counts_0, u_edges_0, v_edges_0 = grid_cube_all_stats(
    frequencies=freq, uu=uu, vv=vv, vis_re=vis_re, vis_imag=vis_imag, weight=wt,
    npix = 501, window_name="kaiser_bessel",
    window_kwargs={"m": 6}, std_min_effective=std_min_effective  # only KB will use beta; m/normalize optional
)
```

```python
def compare_std_gridders_sorted_by_counts(
    std_old, std_new, counts, fallback_mask=None, *,
    channel=0, title="Re", eps=1e-30, max_points=200000
):
    """
    Compare two std (SE) grids vs counts/bin, sorted by counts.
    std_old, std_new, counts: arrays with shape (F, ..., ...)
    fallback_mask optional for highlighting hybrid fallback pixels.
    """
    s_old = np.ravel(std_old[channel])
    s_new = np.ravel(std_new[channel])
    cnt   = np.ravel(counts[channel])

    valid = np.isfinite(s_old) & np.isfinite(s_new) & np.isfinite(cnt) & (cnt > 0) & (s_old > 0) & (s_new > 0)
    if fallback_mask is not None:
        fb = np.ravel(fallback_mask[channel]) & valid
    else:
        fb = np.zeros_like(valid, dtype=bool)

    if not np.any(valid):
        print("No valid pixels to compare.")
        return

    s_old = s_old[valid]
    s_new = s_new[valid]
    cnt   = cnt[valid]
    fb    = fb[valid]

    # sort by increasing counts
    order = np.argsort(cnt)
    s_old = s_old[order]
    s_new = s_new[order]
    cnt   = cnt[order]
    fb    = fb[order]

    # optional downsample for plotting speed
    if s_old.size > max_points:
        step = int(np.ceil(s_old.size / max_points))
        s_old = s_old[::step]
        s_new = s_new[::step]
        cnt   = cnt[::step]
        fb    = fb[::step]

    ratio = s_new / (s_old + eps)
    fracdiff = (s_new - s_old) / (s_old + eps)

    fig = plt.figure(figsize=(14, 12))

    # 1) std vs sorted pixel index (with counts encoded on x only by order)
    ax1 = plt.subplot(3, 1, 1)
    ax1.plot(np.log10(s_old + eps), lw=1, alpha=0.8, label="Old (expansion)")
    ax1.plot(np.log10(s_new + eps), lw=1, alpha=0.8, label="New (hybrid)")
    ax1.set_ylabel("log10(SE)")
    ax1.axvline(np.argwhere(cnt == std_min_effective-1)[-1][0])
    ax1.set_title(f"{title}: SE comparison sorted by increasing counts/bin (ch={channel})")
    ax1.legend()

    # mark fallback points if provided
    if np.any(fb):
        idx_fb = np.where(fb)[0]
        # sparse overlay to keep plot readable
        if idx_fb.size > 30000:
            idx_fb = idx_fb[:: max(1, idx_fb.size // 30000)]
        ax1.scatter(idx_fb, np.log10(s_new[idx_fb] + eps), s=2, alpha=0.5, label="_nolegend_")

    # 2) ratio vs counts
    ax2 = plt.subplot(3, 1, 2)
    ax2.scatter(cnt[~fb], ratio[~fb], s=5, alpha=0.2, label="Non-fallback")
    if np.any(fb):
        ax2.scatter(cnt[fb], ratio[fb], s=5, alpha=0.4, label="Fallback pixels")
    ax2.set_xscale("log")
    ax2.set_yscale("log")
    ax2.set_xlabel("Counts/bin")
    ax2.set_ylabel("SE_new / SE_old")
    ax2.legend()
    ax2.axvline(std_min_effective-1, color = "red")
    ax2.set_title("Ratio vs counts")

    # 3) fractional difference binned by counts
    ax3 = plt.subplot(3, 1, 3)
    # log-spaced count bins
    cmin = max(1.0, np.nanmin(cnt))
    cmax = np.nanmax(cnt)
    if cmax <= cmin:
        bins = np.array([cmin, cmin + 1])
    else:
        bins = np.unique(np.logspace(np.log10(cmin), np.log10(cmax), 30))
        if bins.size < 2:
            bins = np.array([cmin, cmax])

    xb = []
    med = []
    p16 = []
    p84 = []
    for lo, hi in zip(bins[:-1], bins[1:]):
        m = (cnt >= lo) & (cnt < hi)
        if np.sum(m) < 10:
            continue
        xb.append(np.sqrt(lo * hi))
        med.append(np.nanmedian(fracdiff[m]))
        p16.append(np.nanpercentile(fracdiff[m], 16))
        p84.append(np.nanpercentile(fracdiff[m], 84))

    if xb:
        xb = np.asarray(xb)
        med = np.asarray(med)
        p16 = np.asarray(p16)
        p84 = np.asarray(p84)
        ax3.plot(xb, med, marker="o", lw=1, label="Median frac diff")
        ax3.fill_between(xb, p16, p84, alpha=0.25, label="16-84%")
        ax3.set_xscale("log")
        ax3.axhline(0.0, lw=1)
        ax3.axvline(std_min_effective-1, color = "red")
        ax3.set_xlabel("Counts/bin")
        ax3.set_ylabel("(SE_new - SE_old) / SE_old")
        ax3.legend()
        ax3.set_title("Binned trend vs counts")
    else:
        ax3.text(0.5, 0.5, "Not enough points for binned summary", ha="center", va="center")
        ax3.set_axis_off()

    plt.tight_layout()
    plt.show()

    # quick summary stats
    print(f"[{title} ch={channel}] valid pixels: {s_old.size}")
    print(f"Median(SE_new / SE_old) = {np.nanmedian(ratio):.3g}")
    print(f"16-84% range            = [{np.nanpercentile(ratio,16):.3g}, {np.nanpercentile(ratio,84):.3g}]")
    if np.any(fb):
        print(f"Fallback pixels fraction = {fb.mean():.3%}")
        print(f"Median ratio on fallback = {np.nanmedian(ratio[fb]):.3g}")
        if np.any(~fb):
            print(f"Median ratio off fallback= {np.nanmedian(ratio[~fb]):.3g}")
```

```python
# Compare old vs new, sorted by counts/bin
compare_std_gridders_sorted_by_counts(
    std_old=std_re_0, std_new=std_re, counts=counts,
    channel=40, title="Re"
)
```

### Step 3: Gridding with uncertainty estimation!

In this example, we use VisCube's `grid_cube_all_stats` function, which assumes you want to do UV gridding for a spectral cube and calculate the standard deviation within each grid cell. Other gridding functions in VisCube exist, e.g. `grid_cube_all_stats_wbinned`, which does UVW gridding instead. For an example of how to use `grid_cube_all_stats_wbinned`, see [this tutorial instead](https://viscube.readthedocs.io/en/latest/notebooks/xradio_uvw_continuum.html#step-5-load-extracted-numpy-arrays-into-memory-and-do-hermitian-augmentation) starting from Step 5.

The outputs of `grid_cube_all_stats` are as follows:

`mean_re`: the gridded visibilities, real component

`mean_im`: the gridded visibilities, imaginary component

`std_re`: standard deviations, per grid cell, of the real components of the gridded visibilities

`std_im`: standard deviations, per grid cell, of the imaginary components of the gridded visibilities

`counts`: number of visibilities in each grid cell (can be useful for creating a mask also)

`u_edges, v_edges`: the edges of the grid in U,V coordinates

```python
mean_re, mean_im, std_re, std_im, counts, u_edges, v_edges = grid_cube_all_stats(
    frequencies=freq, uu=uu, vv=vv, vis_re=vis_re, vis_imag=vis_imag, weight=wt,
    npix = 501, window_name="kaiser_bessel",
    window_kwargs={"m": 6},   # only KB will use beta; m/normalize optional
)
```

### Step 4 (OPTIONAL): Visualize the results as a "dirty image" velocity map using SuperMAGE

For documentation on how to use SuperMAGE, please see here: https://supermage.readthedocs.io/en/latest/intro.html.

```python
from supermage.utils.plotting import dirty_cube_tool, velocity_map
from supermage.utils.doppler_velocities import create_velocity_grid_stable
fov_arcseconds = 206265*501/(2*461206.80374266725)
num_pixels_uv = 501
arcseconds_per_pixel = fov_arcseconds / num_pixels_uv
freq_path = "/home/darthbarth/Data/ngc_4697/frequencies_correct.npy"
max_freq_index = 74
freq = (np.load(freq_path)/1e9)[:max_freq_index]
systemic_velocity = 1239.9591200072425
velocities_absolute, _ = create_velocity_grid_stable(freq[0], freq[-1], len(freq))
velocities = velocities_absolute.numpy() - systemic_velocity

# Define the region of interest for the cube (pixels 1000 to 1050)
roi_start, roi_end = 225, 276
fov_pixels = roi_end - roi_start
dirty_cube = dirty_cube_tool(mean_re, mean_im, roi_start, roi_end)
extent = (-1*arcseconds_per_pixel*fov_pixels/2, arcseconds_per_pixel*fov_pixels/2, -1*arcseconds_per_pixel*fov_pixels/2, arcseconds_per_pixel*fov_pixels/2)
```

```python
vel_map = velocity_map(dirty_cube[:, :, :max_freq_index], velocities)
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import animation, rc, colors

plt.style.use('dark_background')
fig, ax1 = plt.subplots(1, 1, figsize = (8, 8))
ax1.set_aspect('equal', adjustable='box')

divnorm=colors.TwoSlopeNorm(vcenter=0., vmin = -400, vmax = 400)
im11 = ax1.imshow(vel_map, cmap = "seismic", norm = divnorm, origin = "lower",  extent=extent)
c_white = matplotlib.colors.colorConverter.to_rgba('white',alpha = 0)
c_black= matplotlib.colors.colorConverter.to_rgba('black',alpha = 1)
cmap_rb = matplotlib.colors.LinearSegmentedColormap.from_list('rb_cmap',[c_black, c_white],256)
im12 = ax1.imshow(np.sum(dirty_cube[:, :, :max_freq_index], axis = 2), cmap = cmap_rb, origin = "lower",  extent=extent)

fig.colorbar(im11, label = "Velocity [km/s]", shrink = 0.8)
ax1.set_xlabel("Arcsec")
ax1.set_ylabel("Arcsec")
ax1.set_title("Velocity Map (Dirty)")
plt.grid()
plt.show()
```
